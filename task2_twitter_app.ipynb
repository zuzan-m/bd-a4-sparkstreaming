```python
import socket
import sys
import requests
import requests_oauthlib
import json
```

```python
ACCESS_TOKEN = '*'
ACCESS_SECRET = '*'
CONSUMER_KEY = '*'
CONSUMER_SECRET = '*'
my_auth = requests_oauthlib.OAuth1(CONSUMER_KEY, CONSUMER_SECRET,ACCESS_TOKEN, ACCESS_SECRET)
```

```python
def get_tweets():
    url = 'https://stream.twitter.com/1.1/statuses/filter.json'
    query_data = [('locations', '-100,0,80,50')]
    query_url = url + '?' + '&'.join([str(t[0]) + '=' + str(t[1]) for t in query_data])
    response = requests.get(query_url, auth=my_auth, stream=True)
    print(query_url, response)
    return response
```

```python
def send_tweets_to_spark(http_resp, tcp_connection):
    
    for line in http_resp.iter_lines():
        
        full_tweet = json.loads(line)
        if  (full_tweet['lang'] == 'en') and ('#' in full_tweet['text']) and (not full_tweet['retweeted']) and ('RT @' not in full_tweet['text']):  # to ignore retweets
            try:
                tweet_text = full_tweet['extended_tweet']['full_text']  # full text if possible
            except:
                tweet_text = full_tweet['text']
            print(tweet_text)
            print ("------------------------------------------")
            tcp_connection.send((tweet_text.lower().replace('\n',' ').replace('‚Äô',' ').replace(',',' ').replace('.',' ') + '\n').encode('latin2','ignore'))  #encoding which removed emojis and leaves polish letters
        else:
            pass
       
```

```python
TCP_IP = "localhost"
TCP_PORT = 9009
conn = None
s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
s.bind((TCP_IP, TCP_PORT))
s.listen(1)
print("Waiting for TCP connection...")
conn, addr = s.accept()
print("Connected... Starting getting tweets.")
resp = get_tweets()
send_tweets_to_spark(resp, conn)
```
